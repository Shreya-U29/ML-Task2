```python
import pandas as pd
ufo=pd.read_csv(r'C:\Users\sarvesh\Desktop\coding week\task2.csv')
ufo
```

```python
ufo.isnull().sum()
```

```python
X=ufo.iloc[:,4].values.reshape(-1,1)
y=ufo.iloc[:,6].values.reshape(-1,1)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python

from sklearn.linear_model import LogisticRegression
clf=LogisticRegression().fit(X_train,y_train)
```

```python
clf.predict(X_test)
```

```python
clf.score(X_test,y_test)
```

```python
clf.score(X_train,y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
fc = RandomForestRegressor().fit(X_train, y_train)
fc.predict(X_test)
```

```python
fc.score(X_train,y_train)
```

```python
fc.score(X_test,y_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
lc= GradientBoostingRegressor().fit(X_train, y_train)
lc.score(X_train,y_train)
```

```python
lc.score(X_test,y_test)
```
